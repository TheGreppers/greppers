## How to identify errors — a field guide

A traceback is not punishment, it's a **map**. Here's how to read it.

### 🟥 Spotting a Syntax Error

> *"Python won't even start my program."*

A syntax error is raised **before any code runs**, so you never see your `print` output — only the error.

**Tells:**
- The traceback says `SyntaxError`, `IndentationError`, or `TabError`.
- A small caret `^` points to the exact character that broke parsing.
- The line number in the message is **always reliable** (parsing didn't get past it).

**Top suspects:**
1. **Missing colon** at the end of `if`, `for`, `while`, `def`, `class`, `else`, `elif`.
2. **Unclosed bracket** — `(`, `[`, or `{` without its match. The error often appears on the *next* line.
3. **Mismatched quotes** — `"hello'` or a stray apostrophe inside a `'...'` string.
4. **Bad indentation** — mixing tabs and spaces, or a block that doesn't line up.
5. **Using a reserved word** as a variable (`for = 5`, `class = "Honda"`).

**How to fix:**
- Start at the line number Python gives you, then look **one line above** it (often the real culprit).
- Match opening brackets to closing brackets — count them.
- If indentation looks fine but Python disagrees, retype the whitespace by hand.

---

### 🟧 Spotting a Runtime Error

> *"It runs for a bit, then explodes."*

Code parsed fine and started running, but hit something it couldn't do.

**Tells:**
- You see *some* of your program's output before the traceback.
- The traceback ends with a named exception: `NameError`, `IndexError`, `KeyError`, `TypeError`, `ValueError`, `ZeroDivisionError`, `AttributeError`, `FileNotFoundError`, ...

**Read the traceback bottom-up:**
1. The **last line** names the exception and gives the message — that's the *what*.
2. Above it, lines marked with `File "...", line N, in <function>` show the *where*. Read from bottom to top to follow how you got there.
3. The bottom-most "File" line is your code's failure point.

**The exception name is a giant hint:**

| Exception | What it usually means |
|---|---|
| `NameError` | Typo in a variable or function name; or used before defined |
| `IndexError` | List/string index past the end (off-by-one is the #1 cause) |
| `KeyError` | Dictionary key doesn't exist |
| `TypeError` | Mixed incompatible types — e.g. `"3" + 3`, or called something that isn't a function |
| `ValueError` | Right type, wrong value — e.g. `int("hello")` |
| `ZeroDivisionError` | You divided by 0 (often because a counter started at 0) |
| `AttributeError` | Called `.something()` on a value that doesn't have it (often a `None`) |

**How to fix:**
- Print the suspicious variable **right before** the failing line. The value is almost always the surprise.
- If the error is in a loop, print the loop variable and the index each iteration.

---

### 🟨 Spotting a Logic Error

> *"It runs perfectly. The answer is just wrong."*

The most dangerous kind, because Python tells you nothing.

**Tells:**
- No exception. No red text. The output simply doesn't match what you expected.
- A test that compares actual output to a known-good answer comes back `False`.

**Top suspects:**
1. **Off-by-one** in a loop bound — `<` vs `<=`, `range(n)` vs `range(n+1)`.
2. **Wrong starting value** — counting from 0 when you meant 1 (or vice versa).
3. **Operator precedence** — `a or b and c` doesn't always do what you think. Add parentheses.
4. **Assignment vs comparison** — using `=` where `==` was meant (Python flags this in `if`, but not always elsewhere).
5. **Mutating while iterating** — modifying a list inside a `for` over that same list.
6. **Integer vs float division** — `5 / 2` is `2.5`, but `5 // 2` is `2`.

**How to fix:**
- Add `print(variable)` checkpoints between steps. Inspect what's *actually* in there.
- Write one **test case** with a known answer (e.g. `assert is_odd(3) == True`).
- Read loops aloud: *"i starts at 0, adds 2, stops when i is greater than 10"* — does that produce odds?
- Use the Python debugger or step through mentally with pencil and paper for a few iterations.

---

### The 30-second triage

When something breaks, ask in order:

1. **Did Python show me any of my own output?**
   - No → **Syntax error**. Read line number, look for missing punctuation.
   - Yes → keep going.
2. **Is there a traceback?**
   - Yes → **Runtime error**. Read the bottom line for the exception name.
   - No → keep going.
3. **Output present, no traceback, but wrong?**
   - **Logic error**. Add prints, write a test.

## Round 2 — Runtime Error (60 sec)

**Goal:** Ask for a letter, then tell the user its position in the alphabet.

The code below *starts* fine but crashes. Read the traceback carefully — the last line names the exception, and the file/line above it points at your bug.

**Identification clue:** you'll see the `input` prompt appear (so the program ran), then a traceback ending in `IndexError`.


```python
# BUG: this raises an error while running. Fix the loop bounds.
alphabetList = list("abcdefghijklmnopqrstuvwxyz")
letter = input("Type one letter (a-z): ").lower()

i = 0
while i <= 26:               # <-- is this the right bound?
    if alphabetList[i] == letter:
        print(f"'{letter}' is letter #{i + 1} of the alphabet")
    i += 1
```

<details><summary>💡 Hint</summary>
The list has 26 items, indexed <code>0..25</code>. Using <code>i &lt;= 26</code> asks for <code>alphabetList[26]</code>, which does not exist. Either change to <code>i &lt; 26</code> or stop the loop once the letter is found.
</details>

## Round 3 — Logic Error (90 sec) 🧠

**Goal:** Print the **odd** numbers from 0 to 10.

This code runs with no error... but the answer is wrong. Find the mistake in the *logic*.

**Identification clue:** no traceback. The `Correct?` line at the bottom will print `False`. That's your test catching the logic error.


```python
# BUG: this prints evens, not odds. Change two things to flip it.
odds = []
i = 0

while i <= 10:
    odds.append(i)
    i += 2

print("My 'odds':", odds)

# Self-check - must print True when you are done:
print("Correct?", odds == [1, 3, 5, 7, 9])
```

<details><summary>💡 Hint</summary>
Odd numbers start at <b>1</b>, not 0. The step (<code>+= 2</code>) is still right. Read the loop aloud: <em>"i starts at 1, adds 2 each time, stops when i &gt; 9"</em>.
</details>

## Round 4 — Boss Battle (90 sec) 🔥

**Goal:** Print numbers from 1–100 that are multiples of 2 **or** 5, each shown **exactly once**.

Two bugs hide here: an **off-by-one** (logic) and a **duplicate** problem (also logic). Both run without crashing.

**Identification clue:** no traceback, but the printed total won't match `60` and the duplicates check will be `True`.


```python
# BUG 1: range does not reach 100.
# BUG 2: numbers divisible by BOTH 2 and 5 (like 10, 20) get added twice.
numbers = []
result = []
i = 1
while i < 100:
    numbers.append(i)
    i += 1

for n in numbers:
    if n % 5 == 0:
        result.append(n)
    if n % 2 == 0:
        result.append(n)

print(result[:15], "...")
print("Total found:", len(result), "(should be 60)")
print("Any duplicates?", len(result) != len(set(result)))
```

<details><summary>💡 Hint</summary>
1. To include 100, loop <code>while i &lt;= 100</code>.<br>
2. Use <b>one</b> <code>if</code> with <code>or</code>:  <code>if n % 2 == 0 or n % 5 == 0:</code>
</details>
