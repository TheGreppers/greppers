## 1. Types of Errors (CRD-2.I.1 — CRD-2.I.4)

CollegeBoard defines **four types of errors** that you need to know for the AP exam:

| Error Type | Essential Knowledge | Definition |
|---|---|---|
| **Logic Error** | CRD-2.I.1 | A mistake in the algorithm or program that causes it to behave incorrectly or unexpectedly. |
| **Syntax Error** | CRD-2.I.2 | A mistake in the program where the rules of the programming language are not followed. |
| **Run-Time Error** | CRD-2.I.3 | A mistake in the program that occurs during the execution of a program. Programming languages define their own run-time errors. |
| **Overflow Error** | CRD-2.I.4 | An error that occurs when a computer attempts to handle a number that is outside of the defined range of values. |

### 1a. Syntax Errors (CRD-2.I.2)

A **syntax error** occurs when the rules of the programming language are not followed. The program will **not run at all** — the error is caught before execution.

> **Exam Tip:** A syntax error means the program produces *no behavior*. You cannot observe unexpected behavior from a program that never runs. CollegeBoard has noted that students commonly confuse syntax errors with logic errors on the exam.

```python
# --- SYNTAX ERROR EXAMPLE ---
# This code has a syntax error: missing closing parenthesis
# Uncomment the line below to see the error

# print("Hello, World!"

# Python will raise: SyntaxError: unexpected EOF while parsing
# The program never executes — no output is produced at all.
```

```python
# --- SYNTAX ERROR: Identify and Correct (CRD-2.I.a, CRD-2.I.b) ---

# BROKEN: This function has multiple syntax errors
# def calculate_average(scores)
#     total = 0
#     for score in scores
#         total += score
#     return total / len(scores)

# CORRECTED: All syntax rules are now followed
def calculate_average(scores):   # added colon
    total = 0
    for score in scores:         # added colon
        total += score
    return total / len(scores)

print("Average:", calculate_average([90, 85, 78, 92, 88]))
```

### 1b. Logic Errors (CRD-2.I.1)

A **logic error** is a mistake in the algorithm or program that causes it to behave **incorrectly or unexpectedly**. The program runs without crashing, but produces the **wrong result**.

> Logic errors are the hardest to find because the program does not tell you something is wrong — you have to compare the **expected output** to the **actual output**.

```python
# --- LOGIC ERROR EXAMPLE: Finding the maximum value in a list ---

# BROKEN: Contains a logic error
def find_max_broken(numbers):
    max_val = 0  # Bug: assumes all numbers are positive
    for num in numbers:
        if num > max_val:
            max_val = num
    return max_val

# This works fine for positive numbers...
print("Positive list:", find_max_broken([3, 7, 2, 9, 5]))

# ...but fails for negative numbers (logic error!)
print("Negative list:", find_max_broken([-3, -7, -2, -9, -5]))
# Expected: -2, Actual: 0  <-- WRONG! 0 was never in the list
```

```python
# --- LOGIC ERROR: Corrected (CRD-2.I.b) ---

def find_max_fixed(numbers):
    max_val = numbers[0]  # Fix: initialize to the first element
    for num in numbers:
        if num > max_val:
            max_val = num
    return max_val

print("Positive list:", find_max_fixed([3, 7, 2, 9, 5]))
print("Negative list:", find_max_fixed([-3, -7, -2, -9, -5]))
# Now correctly returns -2 for the negative list
```

### 1c. Run-Time Errors (CRD-2.I.3)

A **run-time error** is a mistake in the program that occurs **during the execution** of a program. The program starts running but crashes at a specific point. Programming languages define their own run-time errors.

```python
# --- RUN-TIME ERROR EXAMPLES ---

# Example 1: Division by zero
def divide(a, b):
    return a / b

try:
    result = divide(10, 0)
except ZeroDivisionError as e:
    print(f"Run-time error caught: {e}")

# Example 2: Index out of range
my_list = [10, 20, 30]
try:
    print(my_list[5])
except IndexError as e:
    print(f"Run-time error caught: {e}")

# Example 3: Wrong type operation
try:
    result = "hello" + 5
except TypeError as e:
    print(f"Run-time error caught: {e}")
```

```python
# --- RUN-TIME ERROR: Corrected with input validation ---

def safe_divide(a, b):
    if b == 0:
        return "Error: Cannot divide by zero"
    return a / b

def safe_list_access(lst, index):
    if index < 0 or index >= len(lst):
        return f"Error: Index {index} is out of range (list has {len(lst)} items)"
    return lst[index]

print(safe_divide(10, 0))
print(safe_divide(10, 3))
print(safe_list_access([10, 20, 30], 5))
print(safe_list_access([10, 20, 30], 1))
```

### 1d. Overflow Errors (CRD-2.I.4)

An **overflow error** occurs when a computer attempts to handle a number that is **outside of the defined range of values**.

> In many languages, integers have a fixed size (e.g., 32-bit). Python handles big integers natively, but overflow still matters in other contexts like fixed-size arrays and other languages.

```python
# --- OVERFLOW ERROR EXAMPLE ---
import numpy as np

# Simulating overflow with a fixed-size integer (8-bit signed: range -128 to 127)
int8_max = np.int8(127)
print(f"Max value for 8-bit signed integer: {int8_max}")

# Adding 1 to the max value causes overflow — wraps around to -128!
overflowed = np.int8(int8_max + np.int8(1))
print(f"127 + 1 with 8-bit integer: {overflowed}")  # -128 (overflow!)

# --- Why this matters on the AP Exam ---
# A 32-bit signed integer can store values from -2,147,483,648 to 2,147,483,647
# Trying to store 2,147,483,648 would cause an overflow error
print(f"\n32-bit signed integer max: {2**31 - 1:,}")
print(f"Adding 1 would overflow to: {np.int32(2**31 - 1) + np.int32(1):,}")
```

## 2. Effective Ways to Find and Correct Errors (CRD-2.I.5)

CollegeBoard identifies **five effective strategies** for finding and correcting errors:

1. **Test cases** — Run the program with different inputs to check for expected outputs
2. **Hand tracing** — Manually step through the code, tracking variable values on paper
3. **Visualizations** — Use tools that show program state graphically
4. **Debuggers** — Use built-in IDE tools to step through code line by line
5. **Adding extra output statement(s)** — Insert print statements to inspect variable values during execution

### Strategy: Hand Tracing

Hand tracing means manually walking through your code line by line, keeping track of each variable's value. This is especially useful for loops and conditionals.

```python
# --- HAND TRACING EXAMPLE ---
# This function is supposed to count how many even numbers are in a list.
# Let's hand trace it to find the bug.

def count_evens_broken(numbers):
    count = 1  # <-- Bug: should start at 0
    for num in numbers:
        if num % 2 == 0:
            count += 1
    return count

# Hand trace with input [3, 4, 7, 8, 2]:
# -----------------------------------------------
# Step | num | num % 2 == 0 | count
# -----------------------------------------------
# init |     |              | 1        <-- wrong! should be 0
#  1   |  3  |    False     | 1
#  2   |  4  |    True      | 2
#  3   |  7  |    False     | 2
#  4   |  8  |    True      | 3
#  5   |  2  |    True      | 4
# -----------------------------------------------
# Returns 4, but the correct answer is 3.

print("Broken result:", count_evens_broken([3, 4, 7, 8, 2]))  # 4 (wrong!)

# CORRECTED
def count_evens_fixed(numbers):
    count = 0  # Fixed: start at 0
    for num in numbers:
        if num % 2 == 0:
            count += 1
    return count

print("Fixed result:", count_evens_fixed([3, 4, 7, 8, 2]))    # 3 (correct!)
```

### Strategy: Adding Extra Output Statements

Inserting `print()` statements at key points in your code lets you see the values of variables during execution. This helps you pinpoint exactly where things go wrong.

```python
# --- EXTRA OUTPUT STATEMENTS TO DEBUG ---
# This function is supposed to return the sum of squares: 1^2 + 2^2 + ... + n^2
# But it's returning the wrong answer. Let's add print statements to debug.

def sum_of_squares_broken(n):
    total = 0
    for i in range(n):  # Bug: range(n) goes from 0 to n-1, missing n
        total += i * i
        print(f"  DEBUG: i={i}, i*i={i*i}, total={total}")  # extra output
    return total

print("sum_of_squares(4) should be 1+4+9+16 = 30")
print(f"Result: {sum_of_squares_broken(4)}")
# The debug output reveals: i goes 0,1,2,3 — never reaches 4!

print("\n--- CORRECTED ---")

def sum_of_squares_fixed(n):
    total = 0
    for i in range(1, n + 1):  # Fix: start at 1, go through n
        total += i * i
    return total

print(f"Result: {sum_of_squares_fixed(4)}")  # 30 (correct!)
```

## 3. Testing for Correctness (CRD-2.J)

### CRD-2.J.1
> In the development process, testing uses **defined inputs** to ensure that an algorithm or program is producing the expected outcomes. Programmers use the results from testing to **revise** their algorithms or programs.

### CRD-2.J.2
> Defined inputs used to test a program should demonstrate the **different expected outcomes** that are at or just beyond the **extremes (minimum and maximum)** of input data.

### CRD-2.J.3
> **Program requirements** are needed to identify appropriate defined inputs for testing.

This is called **boundary testing** or **edge case testing** — you test at the limits of what your program should handle.

```python
# --- TESTING WITH DEFINED INPUTS (CRD-2.J) ---
# Program requirement: Write a function that assigns a letter grade based on a score (0-100).
# A: 90-100, B: 80-89, C: 70-79, D: 60-69, F: 0-59

def get_grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

# Test cases designed using CRD-2.J.2: test at extremes and boundaries
test_cases = [
    # (input, expected_output, description)
    (100, "A", "Maximum score"),
    (90,  "A", "Boundary: minimum A"),
    (89,  "B", "Boundary: maximum B"),
    (80,  "B", "Boundary: minimum B"),
    (79,  "C", "Boundary: maximum C"),
    (70,  "C", "Boundary: minimum C"),
    (69,  "D", "Boundary: maximum D"),
    (60,  "D", "Boundary: minimum D"),
    (59,  "F", "Boundary: maximum F"),
    (0,   "F", "Minimum score"),
]

print("Testing get_grade():")
print("-" * 50)
all_passed = True
for score, expected, description in test_cases:
    actual = get_grade(score)
    status = "PASS" if actual == expected else "FAIL"
    if status == "FAIL":
        all_passed = False
    print(f"  {status} | score={score:3d} | expected={expected} | got={actual} | {description}")

print("-" * 50)
print(f"Result: {'All tests passed!' if all_passed else 'Some tests FAILED'}")
```

## 4. Putting It All Together: Real-World Debugging Exercise

Below is a program with **multiple errors** (syntax, logic, and run-time). Use the strategies from CRD-2.I.5 to identify and correct each one.

```python
# --- DEBUGGING CHALLENGE ---
# This program manages a student roster and calculates the class average.
# It contains THREE errors. Can you find them all?

# BROKEN VERSION (commented out so the notebook still runs):
# --------------------------------------------------------
# def class_average(roster):
#     total = 0
#     for student in roster:
#         total += student["grade"]
#     average = total / len(roster)    # Run-time error if roster is empty!
#     if average >= 90
#         honor_roll = True            # Syntax error: missing colon
#     else:
#         honor_roll = False
#     return average, honor_roll
#
# students = [
#     {"name": "Alice", "grade": 95},
#     {"name": "Bob", "grade": 87},
#     {"name": "Charlie", "grade": 92},
# ]
# avg, honors = class_average(students)
# print(f"Class average: {avg}")                  # Logic error: should show
# print(f"Honor roll: {honors}")                  # if avg >= 90 but also
# print(f"Empty class: {class_average([])}")      # crashes on empty list

# --------------------------------------------------------
# CORRECTED VERSION with all three errors fixed:
# --------------------------------------------------------

def class_average(roster):
    if len(roster) == 0:                     # Fix 1: handle empty list (run-time error)
        return 0, False
    total = 0
    for student in roster:
        total += student["grade"]
    average = total / len(roster)
    if average >= 90:                        # Fix 2: added colon (syntax error)
        honor_roll = True
    else:
        honor_roll = False
    return average, honor_roll

students = [
    {"name": "Alice", "grade": 95},
    {"name": "Bob", "grade": 87},
    {"name": "Charlie", "grade": 92},
]

avg, honors = class_average(students)
print(f"Class average: {avg:.1f}")
print(f"Honor roll eligible: {honors}")

# Now safely handles an empty roster
empty_avg, empty_honors = class_average([])
print(f"\nEmpty class average: {empty_avg}")
print(f"Empty class honor roll: {empty_honors}")
```
